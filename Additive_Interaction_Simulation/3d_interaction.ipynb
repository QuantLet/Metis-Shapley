```python
import matplotlib.pyplot as plt
from matplotlib import cm
from matplotlib.ticker import LinearLocator
import numpy as np

x = np.linspace(0, 2*np.pi, 200)
y = np.linspace(0, 2*np.pi, 200)
X, Y = np.meshgrid(x,y)

m = lambda x, y: np.sin(x) + np.cos(y)

F = m(X,Y)

# Plot the surface (klappt)
fig = plt.figure()
ax = plt.axes(projection='3d')

ax.plot_surface(X, Y, F, cmap='viridis', edgecolor='none')
ax.set_xlabel('$x_1$')
ax.set_ylabel('$x_2$')
ax.set_zlabel('y')
plt.show()
#plt.savefig('3d_upd.png', transparent=True)
```

```python
# (2)
fig = plt.figure()
ax = plt.axes(projection='3d')


sin, cos = np.meshgrid(np.sin(x),np.cos(y))
ax.plot_surface(sin, cos, F, cmap='viridis', edgecolor='none')
ax.set_xlabel('$f(x_1)$')
ax.set_ylabel('$f(x_2)$')
ax.set_zlabel('y')
plt.show()
#plt.savefig('3d_updd.png', transparent=True)
```

```python
#(3)
m = lambda x, y: np.sin(x) + np.cos(y) + 0.1*np.sin(x)*np.cos(y)

F = m(X,Y)

# Plot the surface (klappt)
fig = plt.figure()
ax = plt.axes(projection='3d')

ax.plot_surface(X, Y, F, cmap='viridis', edgecolor='none')
ax.set_xlabel('$x_1$')
ax.set_ylabel('$x_2$')
ax.set_zlabel('y')
plt.show()
#plt.savefig('3d_upddd.png', transparent=True)
```
