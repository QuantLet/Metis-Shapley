```python
import numpy
import numpy as np
import matplotlib
matplotlib.use('TkAgg')
import matplotlib.pyplot as plt
%matplotlib inline
import keras
import tensorflow as tf
from tensorflow import keras
from tensorflow.keras import layers
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Dense, Activation
#SHAP
import shap
import xgboost



import pandas as pd
from sklearn.datasets import make_regression
```

```python
#Import data
design, dep = shap.datasets.boston()
model = xgboost.XGBRegressor().fit(design, dep)

# explain the model's predictions using SHAP
explainer = shap.Explainer(model)
shap_values = explainer(design)

```

```python
# visualize the first prediction's explanation
#plt.figure(figsize=(6.4,4.8), dpi= 200) #(6.4, 4.8) baseline
shap.plots.waterfall(shap_values[5])
plt.tight_layout()
#plt.savefig('boost_1.png', transparent=True)
```

```python
# visualize the first prediction's explanation with a force plot # screenshot
shap.initjs()
shap.plots.force(shap_values[5])
```

```python
# summarize the effects of all the features
shap.plots.beeswarm(shap_values)
plt.tight_layout()
#plt.savefig('boost_2.png', transparent=True)
```

```python
shap.plots.bar(shap_values)
plt.tight_layout()
#plt.savefig('boost_3.png', transparent=True)
```

```python
# create a dependence scatter plot to show the effect of a single feature across the whole dataset
shap.plots.scatter(shap_values[:,"LSTAT"])
plt.tight_layout()
#plt.savefig('boost_4.png', transparent=True)
```
